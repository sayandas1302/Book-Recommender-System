### Preprocess the Datasets

```python
booksDf = pd.read_csv('./Datasets/Books.csv')
```

```python
# removing the missing values 
booksDf = booksDf.dropna(axis=0)

for col in booksDf.columns:
    print(f"{col} : {booksDf[col].isnull().sum()} missing values")
```

```python
# removing duplicated books, ie. books having same book title but different ISBN
booksDf = booksDf.groupby('Book-Title').first().reset_index()
booksDf
```

```python
# removing html character entities like '&amp;'
booksDf['Book-Title'] = booksDf['Book-Title'].apply(lambda x: html.unescape(x))
booksDf['Publisher'] = booksDf['Publisher'].apply(lambda x: html.unescape(x))
```

```python
# saving the preprocessed data frame
booksDf.to_csv('./Datasets/preProcessedBooksDf.csv', index=False)
```

### Popularity Based Recommendation

```python
ratingsDf = pd.read_csv('./Datasets/Ratings.csv')
booksDf = pd.read_csv('./Datasets/preProcessedBooksDf.csv')
```

```python
booksDf.nunique()

# no duplicated books are there
```

```python
# m = minimum number of votes required to be listed in the popularity list
m = 100

# C = average ratings across all over the report 
C = ratingsDf["Book-Rating"].mean()

# selected books
vote_count = ratingsDf["ISBN"].value_counts()
selected_books = list(vote_count[vote_count>=m].index)

# votes per books
v = ratingsDf.loc[ratingsDf["ISBN"].isin(selected_books)]["ISBN"].value_counts()

# average ratings of the selected books
R = ratingsDf.loc[ratingsDf["ISBN"].isin(selected_books)].groupby("ISBN")["Book-Rating"].mean().reindex(selected_books)

# popularity = (vR + mC)/(v+m)
popularity_score = ((v*R+m*C)/(v+m)).reindex(selected_books).reset_index()
popularity_score.columns = ["ISBN", "popularity"]
popularity_score
```

```python
# merging the average ratings with the popularity dataset
avg_ratings = R.reset_index()
avg_ratings.columns = ["ISBN", "avg_ratings"]
popularity_score = pd.merge(popularity_score, avg_ratings, on="ISBN")

# merging thenumber of votes with the popularity dataset
n_votes = v.reset_index()
n_votes.columns = ["ISBN", "nvotes"]
popularity_score = pd.merge(popularity_score, n_votes, on="ISBN")

popularity_score
```

```python
# books information according to the descending order of the popularity score
popularBooks = pd.merge(popularity_score, booksDf, on="ISBN").sort_values('popularity', ascending=False).reset_index()
popularBooks = popularBooks.drop("index", axis=1)
popularBooks['avg_ratings'] = popularBooks['avg_ratings'].map(lambda x: round(x, 2))
popularBooks.head()
```

```python
popularBooks.shape
```

```python
# saving the popularity df
popularBooks.to_csv('./Datasets/popularBooks.csv', index=False)
```

### Collaborative Filtering Based Recommendation

```python
ratingsDf = pd.read_csv('./Datasets/Ratings.csv')
booksDf = pd.read_csv('./Datasets/preProcessedBooksDf.csv')
```

```python
# merging the ratings and books to get a new dataframe
df = pd.merge(ratingsDf, booksDf, on="ISBN")
df
```

```python
# getting the book titles that has atleast 100 votes
vote_count = df[['Book-Title', 'User-ID']].value_counts().groupby('Book-Title').sum()
freq_read_books = list(vote_count[vote_count>=100].index)
freq_read_books
```

```python
len(freq_read_books)
```

```python
# getting the users who has rated atleast 100 books
book_rated_by_users = df[['Book-Title', 'User-ID']].value_counts().groupby('User-ID').sum()
user_rated_many_books = list(book_rated_by_users[book_rated_by_users>=100].index)
user_rated_many_books
```

```python
len(user_rated_many_books)
```

```python
# filtering the merged data to the seleted users
df = df.loc[df["User-ID"].isin(user_rated_many_books),]

# filtering the merged data to the selected books
df = df.loc[df["Book-Title"].isin(freq_read_books),].reset_index().drop('index', axis=1)
df
```

```python
# getting the matrix of which user rated which movies
rating_matrix = df.pivot(index="Book-Title", columns="User-ID", values="Book-Rating")
rating_matrix = rating_matrix.fillna(0)
rating_matrix
```

```python
# similarity scores between the books according to the cosine similarity
sim_matrix = pd.DataFrame(cosine_similarity(rating_matrix))
sim_matrix
```

```python
# function for the recommendation
def recommend(title, n):
    '''function that recommends n books similar to the book given'''
    # id of the books in the list 
    id_in_the_list = freq_read_books.index(title)
    
    # id of the books similar tp the given book
    id_of_the_similar_books = sim_matrix.loc[id_in_the_list,].sort_values(ascending=False)[1:(n+1)].index
    
    # list of recommended books
    recommended_books = [freq_read_books[x] for x in id_of_the_similar_books]

    # informations about the books
    info = booksDf.loc[booksDf["Book-Title"].isin(recommended_books),].drop(["Image-URL-S", "Image-URL-M"], axis=1)

    return info


recommend('1984', 3)
```
